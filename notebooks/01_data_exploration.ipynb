```python
import pandas as pd
import numpy as np

# ── 1. Reload cleanly 
df = pd.read_csv("../data/raw/hmis-item-2018-19-mn-tn-upto-Jan.csv", encoding="latin1")

# ── 2. Keep only the label columns + Total district columns ───
label_cols = ['Indicator', 'S.No.', 'Parameters', 'Type']

# Total columns = district columns that don't end with [A], [B], [C], [D]
total_cols = [col for col in df.columns 
              if 'District' in col 
              and not any(col.strip().endswith(x) for x in ['[A]', '[B]', '[C]', '[D]'])]

print(f"Label columns: {len(label_cols)}")
print(f"Total (district-level) columns: {len(total_cols)}")
print("\nDistrict total columns found:")
for col in total_cols:
    print(" ", col)
```

```python
# ── 3. Build clean dataframe with only label + total columns ──
df_clean = df[label_cols + total_cols].copy()

# ── 4. Rename district columns — strip the long prefix/suffix ─
rename_map = {}
for col in total_cols:
    # Extract just the district name
    # "District - Chennai - Total [(A+B) or (C+D)]" → "Chennai"
    district_name = col.replace("District - ", "").replace(" - Total [(A+B) or (C+D)]", "").strip()
    rename_map[col] = district_name

df_clean = df_clean.rename(columns=rename_map)

# ── 5. Clean the Parameters column ────────────────────────────
df_clean['Parameters'] = df_clean['Parameters'].str.strip()

# ── 6. Melt from wide → long format ───────────────────────────
# Wide:  each row = 1 indicator, each column = 1 district
# Long:  each row = 1 indicator × 1 district (what we need for charts)
district_names = list(rename_map.values())
district_names = [d for d in district_names if d != '_Tamil Nadu']  # keep TN separate

df_long = df_clean[df_clean['_Tamil Nadu'].notna()].melt(
    id_vars=label_cols,
    value_vars=district_names,
    var_name='District',
    value_name='Value'
)

# ── 7. Convert Value to numeric, force errors to NaN ──────────
df_long['Value'] = pd.to_numeric(df_long['Value'], errors='coerce')

# ── 8. Save cleaned files ──────────────────────────────────────
df_clean.to_csv("../data/processed/hmis_wide_clean.csv", index=False)
df_long.to_csv("../data/processed/hmis_long_clean.csv", index=False)

print("✅ Cleaning done!")
print(f"Wide format shape:  {df_clean.shape}")
print(f"Long format shape:  {df_long.shape}")
print(f"\nDistricts in data ({len(district_names)}):")
print(district_names)
print(f"\nUnique indicators: {df_long['Indicator'].nunique()}")
print(f"\nSample of long format:")
df_long.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the clean long format
df_long = pd.read_csv("../data/processed/hmis_long_clean.csv")

# ── CHART 1: ANC Registrations by District (Bar Chart) ────────
anc_data = df_long[
    (df_long['S.No.'] == 1.1) &
    (df_long['Parameters'].str.contains("Total number of pregnant women registered", na=False))
].sort_values('Value', ascending=True)

fig, ax = plt.subplots(figsize=(12, 8))
bars = ax.barh(anc_data['District'], anc_data['Value'], color='steelblue', edgecolor='white')

# Highlight top 3 in a different colour
top3 = anc_data.nlargest(3, 'Value')['District'].values
for bar, district in zip(bars, anc_data['District']):
    if district in top3:
        bar.set_color('#e05c2a')

ax.set_xlabel("Number of Pregnant Women Registered", fontsize=12)
ax.set_title("ANC Registrations by District — Tamil Nadu (2018-19 up to Jan)", 
             fontsize=14, fontweight='bold', pad=15)
ax.axvline(anc_data['Value'].mean(), color='green', linestyle='--', linewidth=1.5, label=f"State avg: {anc_data['Value'].mean():,.0f}")
ax.legend()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
plt.tight_layout()
plt.savefig("../assets/chart1_anc_registrations.png", dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart 1 saved")

# ── CHART 2: Institutional Deliveries by District ─────────────
delivery_data = df_long[
    (df_long['Indicator'].str.contains("M2", na=False)) &
    (df_long['Parameters'].str.contains("Total deliveries conducted", na=False))
].sort_values('Value', ascending=True)

if len(delivery_data) == 0:
    # Try alternate parameter name
    delivery_data = df_long[
        df_long['Indicator'].str.contains("M2", na=False)
    ].groupby('District')['Value'].sum().reset_index().sort_values('Value', ascending=True)
    y_col = 'District'
    x_col = 'Value'
else:
    y_col = 'District'
    x_col = 'Value'

fig, ax = plt.subplots(figsize=(12, 8))
ax.barh(delivery_data[y_col], delivery_data[x_col], color='teal', edgecolor='white')
ax.set_xlabel("Number of Deliveries", fontsize=12)
ax.set_title("Total Deliveries by District — Tamil Nadu (2018-19 up to Jan)",
             fontsize=14, fontweight='bold', pad=15)
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1000:.0f}K'))
plt.tight_layout()
plt.savefig("../assets/chart2_deliveries.png", dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart 2 saved")

# ── CHART 3: Indicator category breakdown (Pie chart) ─────────
indicator_totals = df_long.groupby('Indicator')['Value'].sum().sort_values(ascending=False).head(8)
# Clean labels
labels = [label.split('[')[1].replace(']','').strip() if '[' in label else label 
          for label in indicator_totals.index]

fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(
    indicator_totals.values,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("Set2", len(labels))
)
for text in texts:
    text.set_fontsize(9)
ax.set_title("Distribution of Health Activity Volume by Category\nTamil Nadu HMIS 2018-19",
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig("../assets/chart3_indicator_breakdown.png", dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart 3 saved")

print("\n🎉 All 3 charts saved to /assets folder!")
```
