```python
import pandas as pd
import numpy as np

# ── 1. Reload cleanly 
df = pd.read_csv("../data/raw/hmis-item-2018-19-mn-tn-upto-Jan.csv", encoding="latin1")

# ── 2. Keep only the label columns + Total district columns ───
label_cols = ['Indicator', 'S.No.', 'Parameters', 'Type']

# Total columns = district columns that don't end with [A], [B], [C], [D]
total_cols = [col for col in df.columns 
              if 'District' in col 
              and not any(col.strip().endswith(x) for x in ['[A]', '[B]', '[C]', '[D]'])]

print(f"Label columns: {len(label_cols)}")
print(f"Total (district-level) columns: {len(total_cols)}")
print("\nDistrict total columns found:")
for col in total_cols:
    print(" ", col)
```

```python
# ── 3. Build clean dataframe with only label + total columns ──
df_clean = df[label_cols + total_cols].copy()

# ── 4. Rename district columns — strip the long prefix/suffix ─
rename_map = {}
for col in total_cols:
    # Extract just the district name
    # "District - Chennai - Total [(A+B) or (C+D)]" → "Chennai"
    district_name = col.replace("District - ", "").replace(" - Total [(A+B) or (C+D)]", "").strip()
    rename_map[col] = district_name

df_clean = df_clean.rename(columns=rename_map)

# ── 5. Clean the Parameters column ────────────────────────────
df_clean['Parameters'] = df_clean['Parameters'].str.strip()

# ── 6. Melt from wide → long format ───────────────────────────
# Wide:  each row = 1 indicator, each column = 1 district
# Long:  each row = 1 indicator × 1 district (what we need for charts)
district_names = list(rename_map.values())
district_names = [d for d in district_names if d != '_Tamil Nadu']  # keep TN separate

df_long = df_clean[df_clean['_Tamil Nadu'].notna()].melt(
    id_vars=label_cols,
    value_vars=district_names,
    var_name='District',
    value_name='Value'
)

# ── 7. Convert Value to numeric, force errors to NaN ──────────
df_long['Value'] = pd.to_numeric(df_long['Value'], errors='coerce')

# ── 8. Save cleaned files ──────────────────────────────────────
df_clean.to_csv("../data/processed/hmis_wide_clean.csv", index=False)
df_long.to_csv("../data/processed/hmis_long_clean.csv", index=False)

print("✅ Cleaning done!")
print(f"Wide format shape:  {df_clean.shape}")
print(f"Long format shape:  {df_long.shape}")
print(f"\nDistricts in data ({len(district_names)}):")
print(district_names)
print(f"\nUnique indicators: {df_long['Indicator'].nunique()}")
print(f"\nSample of long format:")
df_long.head(10)
```

```python
import folium
import json
import pandas as pd

# Load GeoJSON
with open("../data/tamilnadu_districts.geojson", "r") as f:
    tn_geojson = json.load(f)

# Load ANC data
df_long = pd.read_csv("../data/processed/hmis_long_clean.csv")
anc_map = df_long[
    df_long['Parameters'].str.contains("Total number of pregnant women registered for ANC", na=False)
][['District', 'Value']].dropna()

# Build the map
m = folium.Map(location=[11.0, 78.5], zoom_start=7, tiles="CartoDB positron")

folium.Choropleth(
    geo_data=tn_geojson,
    name="ANC Registrations",
    data=anc_map,
    columns=["District", "Value"],
    key_on="feature.properties.DISTRICT",
    fill_color="YlOrRd",
    fill_opacity=0.75,
    line_opacity=0.5,
    legend_name="ANC Registrations (2018-19)",
    nan_fill_color="lightgray"
).add_to(m)

# Add district name tooltips
folium.GeoJson(
    tn_geojson,
    name="District Labels",
    tooltip=folium.GeoJsonTooltip(fields=["DISTRICT"], aliases=["District:"])
).add_to(m)

folium.LayerControl().add_to(m)

# Save map
m.save("../assets/tn_anc_map.html")
print("✅ Map saved to assets/tn_anc_map.html")
m
```

```python
# Add clickable popups with ANC data to the map
m2 = folium.Map(location=[11.0, 78.5], zoom_start=7, tiles="CartoDB positron")

# Convert ANC data to dict for easy lookup
anc_dict = anc_map.set_index('District')['Value'].to_dict()

folium.Choropleth(
    geo_data=tn_geojson,
    name="ANC Registrations",
    data=anc_map,
    columns=["District", "Value"],
    key_on="feature.properties.DISTRICT",
    fill_color="YlOrRd",
    fill_opacity=0.75,
    line_opacity=0.5,
    legend_name="ANC Registrations (2018-19)",
    nan_fill_color="lightgray"
).add_to(m2)

# Add popup with district name + ANC count on click
for feature in tn_geojson['features']:
    district = feature['properties']['DISTRICT']
    value = anc_dict.get(district, 0)
    
    # Get center of the bounding box for marker placement
    coords = feature['geometry']['coordinates'][0]
    lats = [c[1] for c in coords]
    lngs = [c[0] for c in coords]
    center = [sum(lats)/len(lats), sum(lngs)/len(lngs)]
    
    folium.Marker(
        location=center,
        popup=folium.Popup(
            f"<b>{district}</b><br>ANC Registrations: <b>{int(value):,}</b>",
            max_width=200
        ),
        icon=folium.DivIcon(
            html=f'<div style="font-size:8px;color:#333;font-weight:bold;white-space:nowrap;">{district}</div>',
            icon_size=(80, 20),
            icon_anchor=(40, 10)
        )
    ).add_to(m2)

folium.LayerControl().add_to(m2)
m2.save("../assets/tn_anc_map.html")
print("✅ Enhanced map saved!")
m2
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

df_long = pd.read_csv("../data/processed/hmis_long_clean.csv")

# ── Build feature matrix: one row per district 

# 1. ANC Registrations
anc = df_long[
    df_long['Parameters'].str.contains("Total number of pregnant women registered for ANC", na=False)
][['District','Value']].rename(columns={'Value':'ANC_Registrations'})

# 2. Total Deliveries
deliveries = df_long[
    df_long['Indicator'].str.contains("M2", na=False)
].groupby('District')['Value'].sum().reset_index().rename(columns={'Value':'Total_Deliveries'})

# 3. BCG Immunisation
bcg = df_long[
    df_long['Parameters'].str.contains("BCG", na=False)
].groupby('District')['Value'].sum().reset_index().rename(columns={'Value':'BCG_Immunisation'})

# 4. Family Planning
fp = df_long[
    df_long['Indicator'].str.contains("M8|M18", na=False)
].groupby('District')['Value'].sum().reset_index().rename(columns={'Value':'Family_Planning'})

# 5. C-Section deliveries
csec = df_long[
    df_long['Indicator'].str.contains("M3", na=False)
].groupby('District')['Value'].sum().reset_index().rename(columns={'Value':'C_Section'})

# Merge all features
df_features = anc.merge(deliveries, on='District')\
                  .merge(bcg, on='District')\
                  .merge(fp, on='District')\
                  .merge(csec, on='District')\
                  .dropna()

print(f"✅ Feature matrix built: {df_features.shape}")
print(f"Districts included: {len(df_features)}")
print("\nFeature preview:")
df_features.head()
```

```python
# ── Standardise features (important for K-means) 
district_names = df_features['District'].values
X = df_features.drop('District', axis=1).values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ── Find optimal K using Elbow + Silhouette 
inertias = []
silhouettes = []
K_range = range(2, 8)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(K_range, inertias, 'bo-', linewidth=2)
axes[0].set_xlabel('Number of Clusters (K)')
axes[0].set_ylabel('Inertia')
axes[0].set_title('Elbow Method')
axes[0].grid(True, alpha=0.3)

axes[1].plot(K_range, silhouettes, 'ro-', linewidth=2)
axes[1].set_xlabel('Number of Clusters (K)')
axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Score (higher = better)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("../assets/kmeans_elbow.png", dpi=150, bbox_inches='tight')
plt.show()

print("Silhouette scores:")
for k, s in zip(K_range, silhouettes):
    print(f"  K={k}: {s:.3f}")
print(f"\n✅ Best K by silhouette: {K_range[silhouettes.index(max(silhouettes))]}")
```

```python
# ── Apply K=3 clustering 
km_final = KMeans(n_clusters=3, random_state=42, n_init=10)
df_features['Cluster'] = km_final.fit_predict(X_scaled)

# Label clusters by their ANC mean (high/mid/low performer)
cluster_means = df_features.groupby('Cluster')['ANC_Registrations'].mean().sort_values(ascending=False)
label_map = {
    cluster_means.index[0]: 'High Performer',
    cluster_means.index[1]: 'Mid Performer', 
    cluster_means.index[2]: 'Low Performer'
}
df_features['Cluster_Label'] = df_features['Cluster'].map(label_map)

print("=== DISTRICT CLUSTERS ===\n")
for label in ['High Performer', 'Mid Performer', 'Low Performer']:
    districts = df_features[df_features['Cluster_Label'] == label]['District'].tolist()
    print(f"🔵 {label} ({len(districts)} districts):")
    print("  ", ", ".join(districts))
    print()

# ── Plotly scatter: ANC vs Deliveries coloured by cluster ─────
fig = px.scatter(
    df_features,
    x='ANC_Registrations',
    y='Total_Deliveries',
    color='Cluster_Label',
    text='District',
    title='District Health Clusters — Tamil Nadu (K-Means, K=3)',
    labels={
        'ANC_Registrations': 'ANC Registrations',
        'Total_Deliveries': 'Total Deliveries',
        'Cluster_Label': 'Cluster'
    },
    color_discrete_map={
        'High Performer': '#2ecc71',
        'Mid Performer':  '#f39c12',
        'Low Performer':  '#e74c3c'
    },
    size='C_Section',
    size_max=35
)
fig.update_traces(textposition='top center', textfont_size=9)
fig.update_layout(
    height=600,
    plot_bgcolor='white',
    xaxis=dict(showgrid=True, gridcolor='#eee'),
    yaxis=dict(showgrid=True, gridcolor='#eee'),
    title_font_size=16
)
fig.write_html("../assets/plotly_clusters.html")
fig.show()
print("✅ Cluster chart saved!")

# ── Save cluster results 
df_features.to_csv("../data/processed/district_clusters.csv", index=False)
print("✅ Cluster data saved to processed/")
```
